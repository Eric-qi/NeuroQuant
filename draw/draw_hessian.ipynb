```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm
```

```python
font_path = './TIMES.TTF'
font_prop = fm.FontProperties(fname=font_path, size=10, )
    
plt.rcParams['axes.unicode_minus'] = False
fig = plt.figure(figsize=(6, 5), dpi=300)

plt.xticks(fontproperties=font_prop,  size=10)
plt.yticks(fontproperties=font_prop, size=10)
plt.xlabel("value", fontproperties=font_prop)
plt.ylabel("frequency", fontproperties=font_prop)

hessian_param = [   [0.0670, 0.0111, 0.0719, 0.0425, 0.0826, 0.0124, 0.0735],
                    [0.0146, 0.0353, 0.1349, 0.0767, 0.0055, 0.0051, 0.0717],
                    [0.0734, 0.1420, 0.1420, 0.0580, 0.0648, 0.0540, 0.0102],
                    [0.0419, 0.0855, 0.1047, 0.0722, 0.0553, 0.0510, 0.0107],
                    [0.0962, 0.0043, 0.0571, 0.0253, 0.0358, 0.0327, 0.0023],
                    [0.0115, 0.0029, 0.0352, 0.0113, 0.0229, 0.0701, 0.0251],
                    [0.0757, 0.0592, 0.0131, 0.0044, 0.0077, 0.0277, 0.125]]

mask = np.tril(np.ones_like(hessian_param, dtype=bool), k=-1)

ax = sns.heatmap(hessian_param, cmap="YlGnBu", mask=mask, annot=True, cbar= True, linewidths=.5, fmt='.2f', annot_kws={"size": 8})

ax.set_xticks([])
ax.set_yticks([])
ax.tick_params(axis='both', which='both', length=0)

# plt.savefig('./heatmap.png', format='png', dpi=600, bbox_inches='tight')
plt.savefig('./heatmap.pdf', format='pdf', bbox_inches='tight')
```
